```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv('Books.csv')
Users = pd.read_csv('Users.csv')
Ratings  = pd.read_csv('Ratings.csv')
```

```python
books = books[books['Book-Title'] != "The Hitchhiker's Guide to the Galaxy"]
books = books[books['Book-Title'] != "Outlander"]
books = books[books['Book-Title'] != "The Color Purple"]
```

```python
print(books.shape)
print(Ratings.shape)
print(Users.shape)
```

## popularity based Recommendation System

```python
rating_with_name = Ratings.merge(books,on='ISBN')
```

```python
num_rating_df= rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating' :'num_ratings'},inplace=True)
num_rating_df
```

```python
# Group by 'Book-Title' and calculate the mean of 'Book-Rating'
avg_rating_df = rating_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()



# Rename the 'Book-Rating' column to 'avg_ratings'
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)

avg_rating_df['avg_ratings'] = avg_rating_df['avg_ratings'].round(2)

# Display the DataFrame
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df= popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df
```

## Collaborative Filltering

```python
x = rating_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title' , columns = 'User-ID' , values = 'Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse = True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))    
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)


    return data
```

```python
recommend('1984')
```
